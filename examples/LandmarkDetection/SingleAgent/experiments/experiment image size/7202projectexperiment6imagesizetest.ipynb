```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#experiment 6: Image size
#NOTE: Any run with less than 10 distance error values is deemed inconclusive (it will not converge)
#NOTE: Both are on memory size 1e5
#NOTE: Image size 13 13 13 uses 125000 model runs, image size 9 9 9 uses 250000 model runs
experiment6_9 = np.array([4.123105625617661, 5.0, 4.58257569495584, 4.358898943540674, 4.358898943540674, 5.830951894845301, 5.0990195135927845, 4.898979485566356, 4.47213595499958, 3.605551275463989]) #Image size 9 9 9
experiment6_13 = np.array([2.449489742783178, 2.449489742783178, 2.8284271247461903, 3.3166247903554, 3.1622776601683795, 3.0, 2.449489742783178, 2.23606797749979, 2.23606797749979, 4.123105625617661]) #Image size 13 13 13
```

```python
imageSizeAmounts = [9, 13]
experiment5_all = [experiment6_9, experiment6_13]
```

```python
experimentAverages = []
for experiment in experiment5_all:
    experimentLength = len(experiment)
    experimentSum = np.sum(experiment)
    experimentAverage = experimentSum/experimentLength
    experimentAverages.append(experimentAverage)
```

```python
experimentAverages
```

```python
plt.plot(imageSizeAmounts, experimentAverages)
plt.title('Average Distance Error for Varying Amounts of Image Sizes (All dimensions are the same)')
plt.xlabel('Model Runs')
plt.ylabel('Average Distance Error')
```
