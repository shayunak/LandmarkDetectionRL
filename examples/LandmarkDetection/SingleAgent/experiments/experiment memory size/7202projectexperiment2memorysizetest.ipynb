```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#experiment 2: Memory size
#NOTE: Any run with less than 10 distance error values is deemed inconclusive (it will not converge)
experiment2_1e5 = np.array([4.123105625617661, 5.477225575051661, 5.196152422706632, 5.385164807134504, 4.358898943540674, 6.0, 4.898979485566356, 5.0990195135927845, 5.0990195135927845, 5.385164807134504]) #1e5 memory size
experiment2_1e6 = np.array([2.0, 2.8284271247461903, 2.449489742783178, 2.449489742783178, 3.1622776601683795, 2.449489742783178, 2.23606797749979, 2.449489742783178, 2.449489742783178, 1.7320508075688772]) #1e6 memory size
```

```python
memorySizeAmounts = [1e5, 1e6]
experiment2_all = [experiment2_1e5, experiment2_1e6]
```

```python
experimentAverages = []
for experiment in experiment2_all:
    experimentLength = len(experiment)
    experimentSum = np.sum(experiment)
    experimentAverage = experimentSum/experimentLength
    experimentAverages.append(experimentAverage)
```

```python
experimentAverages
```

```python
plt.plot(memorySizeAmounts, experimentAverages)
plt.title('Average Distance Error for Varying Amounts of Memory Size')
plt.xlabel('Memory Size')
plt.ylabel('Average Distance Error')
```
