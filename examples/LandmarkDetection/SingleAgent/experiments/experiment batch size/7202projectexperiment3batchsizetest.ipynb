```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#experiment 3: Batch size
#NOTE: Any run with less than 10 distance error values is deemed inconclusive (it will not converge)
experiment3_32 = np.array([108.37435120913067, 99.54396013822235, 98.31581764904364, 103.86529738079028, 100.74224535913422, 101.17806086301516, -87.8072432236597, -62.374911843172995, -78.95017438737453, -42.28198131426312]) #batch size 32
experiment3_48 = np.array([2.0, 2.8284271247461903, 2.449489742783178, 2.449489742783178, 3.1622776601683795, 2.449489742783178, 2.23606797749979, 2.449489742783178, 2.449489742783178, 1.7320508075688772]) #batch size 48
experiment3_64 = np.array([5.0990195135927845, 4.123105625617661, 4.123105625617661, 2.0, 3.3166247903554, 4.123105625617661, 4.0, 3.1622776601683795, 4.47213595499958, 3.1622776601683795]) #batch size 64
```

```python
batchSizeAmounts = [32, 48, 64]
experiment3_all = [experiment3_32, experiment3_48, experiment3_64]
```

```python
experimentAverages = []
for experiment in experiment3_all:
    experimentLength = len(experiment)
    experimentSum = np.sum(experiment)
    experimentAverage = experimentSum/experimentLength
    experimentAverages.append(experimentAverage)
```

```python
experimentAverages
```

```python
plt.plot(batchSizeAmounts, experimentAverages)
plt.title('Average Distance Error for Varying Amounts of Batch Sizes')
plt.xlabel('Model Runs')
plt.ylabel('Average Distance Error')
```
