```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#experiment 5: Update frequency
#NOTE: Any run with less than 10 distance error values is deemed inconclusive (it will not converge)
#NOTE: Update frequency 10 has 90000 model runs compared to 250000 model runs for update frequency 4
experiment5_4 = np.array([4.123105625617661, 5.0, 4.58257569495584, 4.358898943540674, 4.358898943540674, 5.830951894845301, 5.0990195135927845, 4.898979485566356, 4.47213595499958, 3.605551275463989]) #Update frequency 4
experiment5_10 = np.array([3.605551275463989, 3.1622776601683795, 3.0, 4.358898943540674, 3.0, 3.0, 2.449489742783178, 2.8284271247461903, 3.7416573867739413, 2.8284271247461903]) #Update frequency 10
```

```python
updateFrequencyAmounts = [4, 10]
experiment5_all = [experiment5_4, experiment5_10]
```

```python
experimentAverages = []
for experiment in experiment5_all:
    experimentLength = len(experiment)
    experimentSum = np.sum(experiment)
    experimentAverage = experimentSum/experimentLength
    experimentAverages.append(experimentAverage)
```

```python
experimentAverages
```

```python
plt.plot(updateFrequencyAmounts, experimentAverages)
plt.title('Average Distance Error for Varying Amounts of Update Frequencies')
plt.xlabel('Model Runs')
plt.ylabel('Average Distance Error')
```
